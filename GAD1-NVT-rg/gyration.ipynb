```python
from math import *
import matplotlib.pyplot as plt
import matplotlib._color_data as mcd
import matplotlib.pylab as pl
import numpy as np
```

```python
index = 0
histogram = []
#fig, ax = plt.plot(tight_layout=True)
for i in range(180, 310, 20):
    data = np.loadtxt("GAD1_gyrate_T%dK.xvg" %i, comments=["@", "#"],unpack=True)
    ti = int(2/3*len(data[0]))
    gyration = data[1][ti:]
    end_int = ceil(max(gyration)/0.005)
    begin_int = floor(min(gyration)/0.005)
    nbins= end_int-begin_int
    #y,x,_ = ax.hist(gyration, bins=nbins, range=[begin_int*0.05,end_int*0.05],ec='black',density=True)
    y,x,_ = plt.hist(gyration, bins=nbins,range=[begin_int*0.005,end_int*0.005],ec='black',density=True)
    plt.close()
    histogram.append([x, y])
    index += 1
```

```python
marker = ["o", "v", ">", "D", "p", "X", "s"]
for i in range(len(histogram)):
    every = 1
    if i >= 3:
        every = 3
    
    plt.plot(histogram[i][0][0:-1]+0.025, histogram[i][1],linewidth=2,\
             marker=marker[i], markersize=5, markevery=every, markerfacecolor="None",label='T%dK' %(i*20+180))

plt.xlabel('r$_g$ (nm)')
plt.ylabel('P(r$_g$)')
#plt.legend(ncol=2,frameon=True, fontsize=5, numpoints=1)
plt.legend()
plt.xlim(0.78, 1.06)
plt.savefig("Rg_possibility.pdf", format="pdf")
plt.show()
#plt.close()
```
